# play_audio_in_df

## Intro
This simple jupyter notebook will demonstrate loading a speechloop dataframe and make the audio playable via HTML audio element.


## Prerequistites

 - Install Jupyter notebooks with: `pip install jupyter==1.0.0`
 - Run with: `jupyter notebook`

```python
import pandas as pd
from uuid import uuid4 as uuid4
from IPython.core.display import HTML
```

```python
%%html
<script>

  function toggle_visibility(i)
  {
      if (i.style.display === "none") {
          i.style.display = "block";
      } else {
          i.style.display = "none";
      }
  }

</script>

<style>

    .filenameclass{
        cursor: pointer;
        text-decoration: underline blue;
        text-underline-position: under;
        color: blue;
    }
    audio{
        width: 200px;
    }

</style>
```

```python
# Converting links to html tags

def path_to_audio_html(path):
    myuuid = 'x'+str(uuid4())[:8]
    filename = path.rsplit('/')[-1]
    return f'''
    <div>
        <div class='filenameclass' onclick="toggle_visibility({myuuid})">{filename}</div>
        <div id="{myuuid}" style="display:none;">
            <audio controls>
              <source src="{path}" type="audio/wav" preload="none">
                Your browser does not support the audio element.
            </audio>
        </div>
    </div>
    '''
```

```python

df = pd.read_csv('../speechloop/output/20210801_13-37-52_vs-sp-cq-gg.csv')

def replace_text(row):
    x = row['filename'] 
    x = x.replace('./data/','../speechloop/data/') #does not work with full paths in ff
    return x

def tidy_dataframe(df):
    # remove all superflous columns to make it smaller and easier to view
    df.drop(['vs', 'sp', 'cq', 'gg', 'audio_hash', 'transcript_hash', 'transcript'], axis=1, inplace=True)    
    col_map = {k:k.replace('_corrected', '') for k in df.columns}
    keep_cols = [col_map[k] for k in col_map]
    df.rename(columns=col_map, inplace=True)
    return df

df['filename'] = df.apply(replace_text, axis=1)
df = tidy_dataframe(df)
df = df.sample(5)
```

```python
HTML(df.to_html(escape=False,formatters=dict(filename=path_to_audio_html)))
```
